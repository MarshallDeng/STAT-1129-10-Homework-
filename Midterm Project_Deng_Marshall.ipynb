```python
#read the numbers file 
f = open('C:\\Users\\marsh\\Downloads\\numbers .txt', 'r')
num = f.read()
num1 = num.replace('\n', ',').split(',')
num2 = [int(i) for i in num1]
num2.sort()
print(num2)
```

```python
#find frequency of distinct elements
freq = {}
for i in num2:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
print(freq)
```

```python
#import amazon orders file 

import pandas as pd

az = pd.read_csv('C:\\Users\\marsh\\Downloads\\amazon-orders.csv', skip_blank_lines=True)
az
```

```python
#summary statistics for monthly spending
import datetime
az['order_month'] = pd.DatetimeIndex(az['Order Date']).month
def clean_dollar(x):
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
az['Total Charged'] = az['Total Charged'].apply(clean_dollar).astype('float')
azdf = az.groupby('order_month')['Total Charged'].describe()
azdf
```

```python
#break down how many orders the person placed each month
#to do so, create a bar chart for total number of orders each month 

import numpy as np
import matplotlib.pyplot as plt
import pylab 
from pylab import xticks 
az.hist(column = 'order_month')
plt.title('total numbers of orders each month')
xticks(range(4,8,1))
plt.xlabel('month')
plt.ylabel('number of orders')
plt.grid(False)
plt.show()
```

```python
#break down how much the person spent by month 
#to do so, create a pie chart for monthly total spending 

azsum = az.groupby('order_month')['Total Charged'].sum()
month = ['4', '5', '6', '7']
mycolors=['yellow', 'green', 'blue', 'red'] 

plt.pie(azsum, labels=month, colors=mycolors)
plt.legend(title='month', bbox_to_anchor=(1.50, 1.0), loc='upper left')
plt.title('breakdown of monthly total spending')
plt.show()
```

```python
#determine the relationship between how much the person spent and how much he ordered 
#to do so, create lines to compare average monthly spending and total monthly orders  

azavg = az.groupby('order_month')['Total Charged'].mean()
aztotal = az.groupby('order_month')['Total Charged'].count()
month = ['4', '5', '6', '7']

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(month, azavg, 'g-')
ax2.plot(month, aztotal, 'b-')

ax1.set_xlabel('month')
ax1.set_ylabel('avg. monthly spending', color='g')
ax2.set_ylabel('total monthly order', color='b')
plt.title('average monthly spending vs. total monthly order')

plt.show()
```

```python
#subplots with max and average monthly spending 

month = ['4', '5', '6', '7']
azmax = az.groupby('order_month')['Total Charged'].max()
azavg = az.groupby('order_month')['Total Charged'].mean()

#subplot 1 - max
plt.subplot(1,2,1)
plt.plot(month, azmax, color = 'y')
plt.title('max monthly spending')
plt.xlabel('month')
plt.ylabel('max spending')

#subplot 2 - avg
plt.subplot(1,2,2)
plt.plot(month, azavg, color = 'm')
plt.title('average monthly spending')
plt.xlabel('month')
plt.ylabel('avg spending')

plt.tight_layout()
plt.show()
```
